```python
import json
import pandas as pd
import numpy as np
import geopy
from geopy.distance import geodesic

import folium  #needed for interactive map
from folium.plugins import HeatMap
```

```python
competitors = pd.read_csv('market_research5.csv')
competitors
```

```python
#Ver las agencias en un mapa
kavak_shw = [(19.278630, -99.484617), (19.424953, -99.166439), 
             (19.483682, -99.133187), (19.377280, -99.254569), (19.283918, -99.176300)]
shw_names = ['Lerma', 'Florencia', 'Fortuna', 'Santa_Fe', 'Tlalpan']
folium_map = folium.Map(location=[19.4, -99.3],
                        zoom_start=11,tiles="OpenStreetMap")

#Add showroom locations
for showroom in kavak_shw:
    marker = folium.Marker(location=showroom,radius=9)
    marker.add_to(folium_map)
```

```python
comp_coor = list(zip(competitors['lat'], competitors['lng'], competitors['name']))

for comp in comp_coor:
    marker = folium.Marker(location=comp[:2],
                           popup=comp[2][:5],
                           icon=folium.Icon(color='green'))
    marker.add_to(folium_map)

folium_map
```

```python
#Obtener Distancias a Cada Showroom

for i in range(5):
    competitors['distance_'+shw_names[i]] = pd.Series([geodesic(kavak_shw[i],
                                                                x[:2]).km for x in comp_coor])
competitors
```
